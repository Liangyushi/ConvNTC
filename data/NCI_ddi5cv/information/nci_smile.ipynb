```python
import csv
import os
import os.path as osp
import torch
import numpy as np
import pandas as pd
import pickle 
import gzip

fname = os.path.join('allDrugsmile_nci.csv')
# df = pd.read_csv(self.folder + name + ".csv", sep='\t', header=None)
df = pd.read_csv(fname, header=None)
df
```

```python
df.shape[0]
```

```python
# 安装RDKit库
# !pip install rdkit-pypi

from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.DataStructs import TanimotoSimilarity

sim = torch.zeros(df.shape[0],df.shape[0])
for i in range(df.shape[0]):
    drug1_smiles = df[3][i]  # 示例分子1
    for j in range(df.shape[0]):
        # 定义药物的SMILES
        drug2_smiles = df[3][j]  # 示例分子2
        
        # 生成分子对象
        drug1 = Chem.MolFromSmiles(drug1_smiles)
        drug2 = Chem.MolFromSmiles(drug2_smiles)
        
        # 生成分子指纹（Morgan指纹）
        fp1 = AllChem.GetMorganFingerprintAsBitVect(drug1, 2, nBits=2048)
        fp2 = AllChem.GetMorganFingerprintAsBitVect(drug2, 2, nBits=2048)
        
        # 计算Tanimoto相似性
        similarity = TanimotoSimilarity(fp1, fp2)
        sim[i,j]=similarity
        
        print(f'The similarity between the two drugs is: {similarity}')
```
